# SDO/AIA EUV

```python
ds_path = "./data/sdo/aia"
email = "mgjeon@khu.ac.kr"

start_date = "2024-01-01"
end_date = "2024-01-02"
```

```python
from datetime import datetime, timedelta

t_start = datetime.strptime(start_date, "%Y-%m-%d")
t_end = datetime.strptime(end_date, "%Y-%m-%d")
td = timedelta(hours=12)
date_list = [t_start + i * td for i in range((t_end - t_start) // td)]
date_list
```

```python
from itipy.download.download_sdo_aia_euv import SDOAIAEUVDownloader

downloader = SDOAIAEUVDownloader(ds_path=ds_path, email=email)
```

```python
import warnings; warnings.filterwarnings("ignore")

for d in date_list:
    downloader.downloadDate(d)
```

```python
from pathlib import Path
from sunpy.map import Map

dl = Path(ds_path)

for wl in downloader.wavelengths:
    file = sorted((dl / wl).glob("*.fits"))[0]
    smap = Map(file)
    smap.peek()
```
